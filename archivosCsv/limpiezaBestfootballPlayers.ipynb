Para continuar enriqueciendo la data a partir de los datos suministrados, se propone agregar la columna de la edad acutal al año 2024 tiendo en cuenta la fecha de nacimiento de los jugadores 

```python
#Importar el módulo datetime
from datetime import datetime
```

```python
#Calcular la fecha actual
today = datetime.now()
```

```python
#Calcular la edad basada en el año de nacimiento
Best_Players['Age'] = today.year - Best_Players['Born']
```

```python
#Se confirma que la columna haya sido agregada
Best_Players.head()
```

Ya teniendo la columna creada, para efectos de lectura, se solicitia que la columna 'Age', se ubique al lado de la columna 'born'

```python
#Obtener la lista de columnas
cols = Best_Players.columns.tolist()
```

```python
#Encontrar los indices de 'Born' y 'Age'
born_index = cols.index('Born')
age_index = cols.index('Age')
```

```python
#Mover 'Age' al lado de 'Born'
if age_index > born_index:
    cols.insert(born_index + 1, cols.pop(age_index))
```

```python
#Aplicar el Nuevo Orden de Columnas al DataFrame
Best_Players = Best_Players[cols]
```

```python
#confirmar los cambios aplicados
Best_Players.head()
```

En cuanto a eliminación de columnas, se observa que las columnas 'Index' y 'Unnamed' presentan los mismos valores, por lo cual para efectos de identificación de columnas, se opta por eliminar la columna 'Unnamed'

```python
#Eliminación de la columna 'Unnamed'
Best_Players = Best_Players.drop(columns=['Unnamed: 0'])
```
